```python
import os
import time
import logging
from pathlib import Path
from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler
import shutil

# --- UPDATED CONFIGURATION ---
WATCH_DIRECTORY = Path.home() / "Downloads"
DESTINATION_ROOT = Path.home() / "Organized_Files"

EXTENSION_MAP = {
    # Documents
    ".pdf": "Documents/PDFs",
    ".docx": "Documents/Word",
    ".txt": "Documents/Text",
    # Data & Spreadsheets (NEW)
    ".csv": "Documents/Data",
    ".xlsx": "Documents/Data",
    ".xls": "Documents/Data",
    ".json": "Documents/Data",
    # Images
    ".jpg": "Media/Images",
    ".jpeg": "Media/Images",
    ".png": "Media/Images",
    # Archives
    ".zip": "Archives",
    ".tar": "Archives",
    # Executables
    ".exe": "Programs",
    ".msi": "Programs",
}

logging.basicConfig(level=logging.INFO, 
                    format='%(asctime)s - %(message)s',
                    datefmt='%Y-%m-%d %H:%M:%S')

class MoverHandler(FileSystemEventHandler):
    def on_modified(self, event):
        for child in WATCH_DIRECTORY.iterdir():
            if child.is_dir() or child.suffix.lower() in [".crdownload", ".tmp", ".part"]:
                continue
            self.move_file(child)

    def move_file(self, file_path):
        extension = file_path.suffix.lower()
        if extension in EXTENSION_MAP:
            sub_folder = EXTENSION_MAP[extension]
            destination_dir = DESTINATION_ROOT / sub_folder
            destination_dir.mkdir(parents=True, exist_ok=True)
            
            dest_path = destination_dir / file_path.name
            
            if dest_path.exists():
                count = 1
                while dest_path.exists():
                    dest_path = destination_dir / f"{file_path.stem}_{count}{file_path.suffix}"
                    count += 1
            
            try:
                shutil.move(str(file_path), str(dest_path))
                logging.info(f"Moved: {file_path.name} -> {sub_folder}")
            except Exception as e:
                logging.error(f"Error moving {file_path.name}: {e}")

if __name__ == "__main__":
    if not WATCH_DIRECTORY.exists():
        print(f"Error: {WATCH_DIRECTORY} does not exist.")
        exit()

    event_handler = MoverHandler()
    observer = Observer()
    observer.schedule(event_handler, str(WATCH_DIRECTORY), recursive=False)
    
    print(f"🚀 Sentinel is active! Monitoring for Docs, Images, and Data (Excel/CSV)...")
    
    observer.start()
    try:
        while True:
            time.sleep(10)
    except KeyboardInterrupt:
        observer.stop()
    observer.join()
```
